En este reporte se presenta estadística descriptiva sobre el nivel de felicidad de varios países del mundo para el primer y último año incluido en el set de datos y se realiza una comparación de valores con dos países seleccionados.

Se empieza importando la librería `pandas` para la manipulación de la base de datos de la semana del 2019-08-07. Los datos fueron extraídos de github usando la función `read_csv()` de `pandas`.

```python
import pandas as pd
```

```python
data_felicidad = pd.read_csv("https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv")
```

```python
display(data_felicidad.head())
```

El conjunto de datos tiene 17 variables de varios paises del mundo, se estará usando la variable _escalera_vida_ que es un puntaje que indica percepción de calidad de vida en una escala de 0 a 10, siendo el 10 el máximo puntaje.

Para encontrar el primer y último año en el conjunto de datos se usan las funciones `min()` y `max()` en la columna _anio_. 


```python
#Medida de tendencia central y dispersión de la felicidad a nivel mundial para el primer y último año incluido en el set de datos.
#find min and max year
min_a = data_felicidad['anio'].min()
max_a = data_felicidad['anio'].max()
print("El primer año es: " + str(min_a))
print("El último año es: " + str(max_a))
```

Conociendo el primer y último año de registro del conjunto de datos se usa la información como condicional para filtar el dataframe y presentar un análisis descriptivo de los datos usando la función `describe()`. Estos conjuntos de datos serán utilizados para análisis posteriores.

```python
felicidad_min = data_felicidad[data_felicidad["anio"] == min_a]
print("Datos_2005")
print(felicidad_min["escalera_vida"].describe())
```

En el año 2005 existen 27 países, su media es 6.45 y su desviación estándar es 0.92. Es importante notar que el puntaje más bajo es 4.72 y el más alto 8.02.

```python
felicidad_max = data_felicidad[data_felicidad["anio"] == max_a]
print("Datos_2018")
print(felicidad_max["escalera_vida"].describe())
```

En comparación con el año 2005, la cantidad de países de los que se tiene información es mayor siendo 136 en total con un promedio de 5.50 y desviación estándar de 1.10. Adicionalmente, el mínimo y el máximo descendieron siendo, en el año 2008, 2.69 y 7.85 respectivamente, a su vez la media también descendió aproximandamente un punto.   

Los países que fueron seleccionados para realizar el estudio comparativo son: Venezuela y Australia, y fueron escogidos arbitrariamente. 

Para el año 2005 tenemos los siguientes puntajes en ambos países.

```python
Venezuela_2005 = felicidad_min[felicidad_min["pais"] == "Venezuela"]
print("La felicidad en Venezuela 2005 es: " + Venezuela_2005["escalera_vida"].to_string(index=False))
Australia_2005 = felicidad_min[felicidad_min["pais"] == "Australia"]
print("La felicidad en Australia 2005 es: " + Australia_2005["escalera_vida"].to_string(index=False))
```

Los puntajes de felicidad de Venezuela y Australia en el año 2005 están por encima de la media del conjunto de datos y son bastante similares entre sí.

```python
Venezuela_2018 = felicidad_max[felicidad_max["pais"] == "Venezuela"]
print("La felicidad en Venezuela 2018 es: " + Venezuela_2018["escalera_vida"].to_string(index=False))
Australia_2018 = felicidad_max[felicidad_max["pais"] == "Australia"]
print("La felicidad en Australia 2018 es: " + Australia_2018["escalera_vida"].to_string(index=False))
```

Para el año 2018 el puntaje de felicidad en Venezuela desciende aproximadamente 2 puntos y ahora se encuentra por debajo de la media. Por otro lado, Australia mantiene su puntaje y está muy cercano al valor máximo en ese año.

Para conocer brevemente las posibles variables que pueden influir en el puntaje de felicidad, se realizó un agrupamiento por país en la data de los años previamente mencionados tomando en consideración la información de las columnas _percepcion_corrupcion_, _soporte_social_ y _confianza_.

La variable soporte social es el promedio nacional de las respuestas binarias 0 o 1 que indica si las personas tienen alguien con quien contar en situaciones difíciles, la variable percepción de la corrupción es el promedio nacional entre las respuestas binarias de si consideran que la corrupción esta extendida a nivel gubernamental y la variable confianza es el promedio nacional de las respuestas binarias de si confian o no en el gobierno. 

Se uso la función `groupby()` para el agrupamiento y las función `mean()` para conocer el promedio de las variables en cada país.

```python
mean_2005 = felicidad_min.groupby('pais')[["escalera_vida", "percepcion_corrupcion", "soporte_social", "confianza"]].mean()
print("Promedio 2005:")
display(mean_2005)
```

En el año 2005 la percepción de la corrupción en Venezuela era bastante alta en comparación a Dinamarca que es el país con mayor puntaje de felicidad, de igual forma la confianza en el gobierno es menor en Venezuela. Sin embargo, el promedio de soporte social es similar en ambos países.

```python
mean_2018 = felicidad_max.groupby('pais')[["escalera_vida", "percepcion_corrupcion", "soporte_social", "confianza"]].mean()
print("Promedio 2018:")
pd.set_option('display.max_rows', None)
display(mean_2018)
```

Para el año 2018 el país con mayor puntaje de felicidad es Finlandia que tiene un puntaje bajo en percepción de la corrupción y un nivel medio en confianza en el gobierno, de igual manera Australia tiene niveles similares a Finlandia en las 3 variables analisadas. Venezuela en cambio tiene un nivel muy alto en percepción de la corrupción y muy bajo en la confianza en al gobierno, además el soporte social descendió aproximademente 0.7 en los 13 años de diferencia que hay entre 2005 y 2018. 

A niveles generales se puede concluir que la similitud principal que existe en los países con un puntaje bajo de felicidad es que desconfían del gobierno y consideran que la corrupción esta impregnada en el sistema político. Por lo que se puede concluir que las decisiones gubernamentales infieren en la calidad de vida del pueblo que gobiernan.
